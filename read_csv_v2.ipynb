```python
from langchain_experimental.agents import create_pandas_dataframe_agent
```

```python
from langchain.llms import VertexAI
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/test_faulty_data.csv")
```

```python
pd.set_option('display.max_columns', None)
df.head(11)
```

```python
llm = VertexAI(model_name="gemini-pro", temperature=0)
```

```python
p_agent = create_pandas_dataframe_agent(llm=llm,df=df,verbose=True,handle_parsing_errors=True)
```

```python
p_agent.run("Find the datatype of the columns and if some content of the column do not match the datatype")
```
